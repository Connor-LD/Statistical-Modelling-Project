```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import sqlite3
from sqlite3 import Error
```

Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
stations = pd.read_csv('stations.csv')
stations.head(3)
```

```python
station_venues_yelp = pd.read_csv('station_venues_yelp.csv')
station_venues_yelp.head(3)
# station_venues_yelp.shape
```

```python
#create a simple linear model to express the popularity of each point of interest, crudely as rating*reviews
station_venues_yelp['poi_popularity'] = station_venues_yelp['poi_rating']*station_venues_yelp['poi_review_count']
station_venues_yelp.head(3)
```

```python
# station_venues_yelp.drop(columns='Unnamed: 0')
```

```python
station_popularity = pd.DataFrame(station_venues_yelp.groupby('station_name')['poi_popularity'].sum().reset_index())
station_popularity.columns = ['station_name', 'poi_popularity_sum'] 
station_popularity
```

```python
station_avgs = pd.DataFrame(station_venues_yelp.groupby('station_name').mean().reset_index())
station_avgs.columns = ['station_name', 'unnamed','station_lat','station_lng','avg_poi_rating', 'avg_poi_review_count', 'avg_poi_price', 'avg_poi_distance', 'avg_poi_popularity']
station_avgs
```

```python
stations_joined = pd.concat([station_avgs, station_popularity, stations], axis=1)
```

```python
stations_joined.columns
```

```python
stations_joined = stations_joined.drop(columns=['unnamed', 'station_lat', 'station_lng','Unnamed: 0','empty_slots', 'free_bikes', 'timestamp', 'extra.ebikes', 'extra.has_ebikes','extra.last_updated', 'extra.normal_bikes', 'extra.renting','extra.returning', 'extra.slots', 'extra.uid','station_name'])
```

```python
stations_joined = stations_joined.dropna()
```

```python
# 35 stations have 0 POIs with reviews from yelp. Excluded from analysis with concession that this analysis cannot represent all stations in vancouver.
stations_joined.info()
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
# check correlation between variables
corrmat = stations_joined.corr()
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(corrmat, vmax=.8, square=True,cmap="RdYlGn_r")
```
