# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

&& Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
# Bike stations from csv created in part 1
stations = pd.read_csv('stations.csv')
stations.head(3)
```

```python
fsq_api_key = os.getenv('FOURSQUARE_API_KEY')
```

```python
# Request Test

# params
radius = 1000
max_price = 3
limit = 10
FSQ_KEY=fsq_api_key
        
# Part 1: creating the API request URL
url = "https://api.foursquare.com/v3/places/search?ll={},{}&radius={}&max_price={}&limit={}".format(
            48.428023,
            -71.062506,
            # lat,
            # lng,
            radius,
            max_price,
            limit
        )

headers = {
    "Authorization": fsq_api_key,
    "accept": "application/json"
        }
           
# Part 2: making the GET request

fsq_response = requests.request("GET", url, headers=headers).json()
```

```python
JSON(fsq_response)
```

```python
# Define function

def getFSQ(names, latitudes, longitudes):
    
# params    
    radius = 240   ## 1 city block ~80m, so 240m is ~9 square blocks
    max_price = 3
    limit = 50

# Instantiate
    venues_list=[]
    for name, lat, lng in zip(stations['name'], stations['latitude'], stations['longitude']):
        print(name)
        
# Part 1: creating the API request URL
        url = "https://api.foursquare.com/v3/places/search?ll={},{}&radius={}&max_price={}&limit={}".format(
            lat,
            lng,
            radius,
            max_price,
            limit
        )

        headers = {
            "Accept": "application/json",
            "Authorization": "fsq3sRrgu5QyANimkXCYi9FbzDRt/xyTx1arDpH/JRBZSr4="
            ##### "Authorization": '"$FOURSQUARE_API_KEY"'
        }
           
# Part 2: making the GET request

        fsq_response = requests.request("GET", url, headers=headers).json()['results']
        
# Part 3: returning only relevant information for each nearby venue and append to the list 
        venues_list.append([(
            name,
            lat, 
            lng,
            v['fsq_id'],
            v['name'],
            v['categories'][0]['id'],
            v['categories'][0]['name'],
            v['distance']) for v in fsq_response])
        
    return (venues_list)
```

```python
station_venues_fsq = getFSQ(names = stations['name'],
                            latitudes = stations['latitude'],
                            longitudes = stations['longitude'])
```

```python
station_venues_fsq
```

Put your parsed results into a DataFrame

```python
station_venues_fsq = pd.DataFrame([item for venue_list in station_venues_fsq for item in venue_list])
```

```python
station_venues_fsq.head(3)
```

```python
station_venues_fsq.columns = ['station_name', 
                          'station_latitude', 
                          'station_longitude', 
                          'fsq_id', 
                          'poi_name', 
                          'poi_id', 
                          'poi_category',
                          'poi_distance']
```

```python
station_venues_fsq.head(5)
```

```python
station_venues_fsq.info()
```

```python
station_venues_fsq.describe()
```

```python
station_venues_fsq.to_csv('station_venues_fsq.csv')
```

# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python
yelp_api_key = os.getenv('YELP_API_KEY')
```

```python
# Request Test

# params
radius = 1000
price1 = 1
price2 = 2
price3 = 3
limit = 50
YELP_KEY=yelp_api_key
        
# Part 1: creating the API request URL
url = "https://api.yelp.com/v3/businesses/search?latitude={}&longitude={}&radius={}&price={}&price={}&price={}&sort_by=best_match&limit={}".format(
            48.428023,   #lat
            -71.062506,  #lng
            radius,
            price1,
            price2,
            price3,
            limit
        )

headers = {
    "Authorization": "Bearer {}".format(yelp_api_key),
    "accept": "application/json"
        }
           
# Part 2: making the GET request

yelp_response = requests.request("GET", url, headers=headers).json()
```

```python
JSON(yelp_response)
```

```python
# Define function

def getYelp(names, latitudes, longitudes):
    
# params    
    radius = 240   ## 1 city block ~80m, so 240m is ~9 square blocks
    price1 = 1
    price2 = 2
    price3 = 3
    limit = 20

# Instantiate
    venues_list=[]
    for name, lat, lng in zip(stations['name'], stations['latitude'], stations['longitude']):
        print(name)
        
# Part 1: creating the API request URL
        url = "https://api.yelp.com/v3/businesses/search?latitude={}&longitude={}&radius={}&price={}&price={}&price={}&sort_by=best_match&limit={}".format(
            lat,
            lng,
            radius,
            price1,
            price2,
            price3,
            limit
        )

        headers = {
            "Authorization": "Bearer {}".format(yelp_api_key),
            "accept": "application/json"
            ##### "Authorization": '"$YELP_API_KEY"'
        }
           
# Part 2: making the GET request
        yelp_response = requests.request("GET", url, headers=headers).json()['businesses']
        
# Part 3: returning only relevant information for each nearby venue and append to the list 
        venues_list.append([(
            name, 
            lat, 
            lng,
            v['id'],
            v['name'],
            v['categories'][0]['title'],
            v['rating'],
            v['review_count'],
            v['price'],
            v['distance']) for v in yelp_response])
        
    return (venues_list)
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
station_venues_yelp = getYelp(names = stations['name'],
                            latitudes = stations['latitude'],
                            longitudes = stations['longitude'])
```

Put your parsed results into a DataFrame

```python
station_venues_yelp
```

```python
station_venues_yelp = pd.DataFrame([item for venue_list in station_venues_yelp for item in venue_list])
```

```python
station_venues_yelp.head(3)
```

```python
station_venues_yelp.columns = ['station_name',
                               'station_latitude', 
                               'station_longitude', 
                               'yelp_id', 
                               'poi_name', 
                               'poi_title', 
                               'poi_rating',
                               'poi_review_count',
                               'poi_price',
                               'poi_distance']
```

```python
station_venues_yelp.head(5)
```

```python
station_venues_yelp['poi_price']=station_venues_yelp['poi_price'].str.len()
```

# Comparing Results

Which API provided you with more complete data? Provide an explanation. 

## Answer

Yelp is the more valuable dataset because it contains ratings, price, and number of reviews. 
However, yelp only contains businesses with reviews, so the list of places is larger with FourSquare (4k vs 2k). Additional differences may be result of differences in arbitrary classification of 'price' level. 
FourSquare also came with categories, but these were often so different that they were difficult to rely on.
